```python
def create_delta_temperature(df):
    df['delta_temperature'] = df['temperature'].diff()
    return df
```

**Task 3**

```python
def plot_dfs(df):
    # Using Matplotlib
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    plt.plot(df['temperature'])
    plt.title('Temperature')
    plt.subplot(2, 1, 2)
    plt.plot(df['delta_temperature'])
    plt.title('Delta Temperature')
    plt.tight_layout()
    plt.show()
    # Using hvplot
    df.hvplot.line(y='temperature', width=800, height=300, xlabel='Index', ylabel='Temperature', title='Temperature') * \
    df.hvplot.line(y='delta_temperature', width=800, height=300, xlabel='Index', ylabel='Delta Temperature', title='Delta Temperature')
```

```python
def plot_temperature_color(df):
    plt.figure(figsize=(12, 6))
    colors = np.where(df['temperature'] > 70.0, 'red', 'blue')
    plt.scatter(df.index, df['temperature'], c=colors)
    plt.xlabel('Index')
    plt.ylabel('Temperature')
    plt.title('Temperature (Color-coded)')
    plt.show()
```

Task 6

```python
from sklearn.impute import SimpleImputer

def rmse_score(df_missing, df_original):
    # Perform missing value imputation on df_missing
    imputer = SimpleImputer(strategy='mean')
    df_missing_imputed = imputer.fit_transform(df_missing[['temperature']])
    
    # Calculate RMSE score
    from sklearn.metrics import mean_squared_error
    rmse = np.sqrt(mean_squared_error(df_missing_imputed, df_original['value']))
    return rmse
```

**Task 7**

```python
def resample_hourly(df):
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df = df.set_index('timestamp')
    df_hourly = df.resample('H').mean()
    return df_hourly
```

```python
def iqr_outliers(df):
    Q1 = df['temperature'].quantile(0.25)
    Q3 = df['temperature'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df['temperature'] < lower_bound) | (df['temperature'] > upper_bound)]
    return outliers
```
